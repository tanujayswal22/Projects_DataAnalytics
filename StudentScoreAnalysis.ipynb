```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("students_score.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

Drop unnamed column

```python
df=df.drop("Unnamed: 0",axis=1)
print(df.head())
```

```python
#change weekly study hours column
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

Gender Distribution

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender") 
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#from the above chart we have analysed that:
#the number of females in the data is more than the number of males
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb, annot=True)
plt.title("Relationship btn Parent's Education and Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have a good impact on their scores
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1, annot=True)
plt.title("Relationship btn Parent's Marital status and Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that there is no/negligible ompact on the
# student's score due to their parents marital status
```

Distribution of Ethnic Groups

```python
groupA=df.loc[(df["EthnicGroup"]=="group A")].count()
groupB=df.loc[(df["EthnicGroup"]=="group B")].count()
groupC=df.loc[(df["EthnicGroup"]=="group C")].count()
groupD=df.loc[(df["EthnicGroup"]=="group D")].count()
groupE=df.loc[(df["EthnicGroup"]=="group E")].count()
l=["groupA","groupB","groupC","groupD","groupE"]
mylist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(mylist)
plt.pie(mylist,labels= l,autopct="%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
